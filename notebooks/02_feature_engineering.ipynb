# Feature Engineering
In this notebook, we will:
- Parse JSON-like fields.
- Extract relevant information from cast and crew.
- Clean and combine text-based features.
- Create the `tags` column that represents each movie.

```python
import pandas as pd
import ast
```

## Load Preprocessed Data

```python
movies = pd.read_csv('../data/clean_movies.csv')
movies.head()
```

## Parsing JSON-like Columns

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L

movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

## Extracting Key Cast and Director

```python
def convert_cast(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter < 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L

def get_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L

movies['cast'] = movies['cast'].apply(convert_cast)
movies['crew'] = movies['crew'].apply(get_director)
movies.head()
```

## Text Normalization

```python
def concat(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1

movies['genres'] = movies['genres'].apply(concat)
movies['keywords'] = movies['keywords'].apply(concat)
movies['cast'] = movies['cast'].apply(concat)
movies['crew'] = movies['crew'].apply(concat)
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head()
```

## Creating the `tags` Feature

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
new_movies = movies.drop(columns=['overview','genres','keywords','cast','crew'])

# Join lists into a single string
new_movies['tags'] = new_movies['tags'].apply(lambda x: " ".join(x))
new_movies.head()
```

## Saving Feature Data

```python
new_movies.to_csv('../data/tags_movies.csv', index=False)
```
